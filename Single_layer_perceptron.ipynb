# Data preprocessing-Applying minmax scaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data[data.columns[0:57]]=scaler.fit_transform(data[data.columns[0:57]])
```

# Splitting data into x and y

```python
x = data[data.columns[0:57]].values
y=data[data.columns[57]]
```

```python
x
```

```python
y
```

# Train-test split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20,random_state=101)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

# Perceptron model

```python
from sklearn.linear_model import Perceptron
p = Perceptron(eta0=1.0,max_iter=100)
p.fit(x_train, y_train)
y_pred=p.predict(x_test)
from sklearn.metrics import accuracy_score
perc_acc=accuracy_score(y_test, y_pred)
print("accuracy= ",perc_acc)
```

# Model evaluation

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold

# define model evaluation method
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate model
scores = cross_val_score(p, x, y, scoring='accuracy', cv=cv, n_jobs=-1)

from numpy import mean
from numpy import std
# summarize result
print('Mean Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score


recall_perc=recall_score(y_test, y_pred, average='macro')
print("recall score= ",recall_perc)

precision_perc=precision_score(y_test, y_pred, average='macro')
print("precision score= ",precision_perc)
```
